```python
import random

def monte_carlo_pi(convergence_criterion, draw_limit):
    inside_circle = 0
    total_draws = 0

    while True:
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)

        distance = x**2 + y**2

        if distance <= 1:
            inside_circle += 1

        total_draws += 1

        current_estimate = (inside_circle / total_draws) * 4
        percentage_difference = abs((current_estimate - 3.14159265358979323846) / 3.14159265358979323846)

        if percentage_difference < convergence_criterion:
            print(f"Convergence achieved within {convergence_criterion} percentage difference.")
            break

        if total_draws >= draw_limit:
            print(f"Draw limit exceeded ({draw_limit} draws). Percentage difference: {percentage_difference}")
            break

    return total_draws, current_estimate

# Prompt the user for input values
convergence_criterion = float(input("Enter the convergence criterion (e.g., 0.01): "))
draw_limit = int(input("Enter the draw limit: "))

# Run Monte Carlo simulation
draws, estimate = monte_carlo_pi(convergence_criterion, draw_limit)

# Print results
print(f"\nMonte Carlo simulation results:")
print(f"Estimated Pi: {estimate}")
print(f"Number of draws: {draws}")
```

```python
import numpy as np

# Define convergence criteria
convergence_criteria = [0.01, 0.001, 0.0001, 0.00001]

# Number of runs for each criterion
num_runs = 10

# Lists to store results
average_draws_list = []
std_draws_list = []

# Run experiments
for convergence_criterion in convergence_criteria:
    draws_list = []

    for _ in range(num_runs):
        draws, _ = monte_carlo_pi(convergence_criterion, draw_limit=float('inf'))
        draws_list.append(draws)

    average_draws = np.mean(draws_list)
    std_draws = np.std(draws_list)

    average_draws_list.append(average_draws)
    std_draws_list.append(std_draws)

# Print statistics
print("\nExperiment Results:")
for i, convergence_criterion in enumerate(convergence_criteria):
    print(f"\nConvergence Criterion: {convergence_criterion}")
    print(f"Average Number of Draws: {average_draws_list[i]}")
    print(f"Standard Deviation of Draws: {std_draws_list[i]}")
```
